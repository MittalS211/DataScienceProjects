```python
import sqlite3
import pandas as pd
```

```python
#What are the three longest trips on rainy days?
pd.read_sql("""
WITH
    rainy_days
AS (
    -- A simple query to get rainy days.
    SELECT
        Date
    FROM
        weather
    WHERE 
        Events='Rain'
)

-- Joining the rainy days table we created with the trips table to get longest days.
SELECT
    t.trip_id, t.duration trip_length
FROM
    trips t
LEFT JOIN
    rainy_days
ON
    t.start_date = rainy_days.Date
ORDER BY 
    trip_length DESC
LIMIT 3
""", conn)
```

```python
#which station is full most often.
pd.read_sql("""SELECT
                  status.station_id,
                  stations.name,
                  COUNT(CASE WHEN docks_available=0 then 1 END) empty_count
               FROM status
               JOIN stations
               ON stations.station_id = status.station_id
               GROUP BY 1
               ORDER BY empty_count DESC
               LIMIT 1
             """, conn)
```

```python
pd.read_sql_query("""SELECT t.start_station, 
                            COUNT(t.trip_id) Trips
                     FROM trips t
                     JOIN stations s
                     ON t.start_station = s.name
                     GROUP BY 1
                     ORDER BY dockcount DESC""", conn)
```

```python
#  What's the length of the longest trip for each day it rains anywhere?
#What are the three longest trips on rainy days?
pd.read_sql("""
WITH
    rainy_days
AS (
    -- A simple query to get rainy days.
    SELECT
        Date
    FROM
        weather
    WHERE 
        Events='Rain'
)

-- Joining the rainy days table we created with the trips table to get longest days.
SELECT
    Date(t.start_date) trip_date, max(t.duration) longest_trip
FROM
    trips t
LEFT JOIN
    rainy_days
ON
    t.start_date = rainy_days.Date
GROUP BY trip_date
ORDER BY 
    trip_date, longest_trip DESC
""", conn)
```
