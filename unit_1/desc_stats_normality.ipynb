```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
weibull=np.random.weibull(1, 1000)
```

```python
plt.hist(weibull)
# Add a vertical line at the mean.
plt.axvline(weibull.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(weibull.mean() + weibull.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(weibull.mean()-weibull.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
power = np.random.power(5, 1000)
```

```python
plt.hist(power)
# Add a vertical line at the mean.
plt.axvline(power.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(power.mean() + power.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(power.mean()-power.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
neg_binomial=np.random.negative_binomial(1, .5, 10000)
```

```python
plt.hist(neg_binomial)
# Add a vertical line at the mean.
plt.axvline(neg_binomial.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(neg_binomial.mean() + neg_binomial.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(neg_binomial.mean()-neg_binomial.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

## Obervations:
Since Weibull, power, logseries and negative_binomial ditributions all have high degrees of skew present in them, measures of central tendency are unhelpful. The mean might suggest a value at which there is a greater density of similar values, but does it doesn't suggest the general direction of the skew present in the distribution.

```python
uni=np.random.uniform(1, 2,10000)
```

```python
plt.hist(uni)
# Add a vertical line at the mean.
plt.axvline(uni.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(uni.mean() + uni.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(uni.mean()-uni.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

Unsurprisingly, measures of central tendency and variance are quite useless for a uniform distribution. As the histogrogram shows, values are fairly evenly distributed in the given range of 1.0 to 2.0, occurring with almost uniform frequency
