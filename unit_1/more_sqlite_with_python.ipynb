```python
import sqlite3
import numpy as np
import pandas as pd
```

```python
conn = sqlite3.connect('/Users/markespina/Downloads/bikeshare.db')

cur = conn.cursor()
```

```python
#What was the hottest day in our data set? Where was that?
pd.read_sql_query("""SELECT MAX(MaxTemperatureF), 
                            Date, 
                            ZIP
                     FROM weather
                     """, conn)
```

```python
#How many trips started at each station?

pd.read_sql_query("""SELECT start_station, 
                            COUNT(trip_id) Trips
                     FROM trips
                     GROUP BY start_station
                     ORDER BY 2 DESC""", conn)
```

```python
#What's the shortest trip that happened
pd.read_sql_query("""SELECT trip_id, 
                            start_station, 
                            end_station, 
                            MIN(duration)
                     FROM trips""", conn)
```

```python
pd.read_sql_query("""SELECT end_station, 
                            AVG(duration) average_duration
                     FROM trips
                     GROUP BY 1
                     ORDER BY 2""", conn)
```
